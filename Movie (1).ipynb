```python
# remove unused data columns in the analysis from the dataset
df.drop(['imdb_id', 'homepage', 'tagline', 'overview', 'budget_adj', 'revenue_adj'], axis =1, inplace =  True)
```

QUESTION 1: WHICH MOVIES GENERATES THE MOST PROFIT?

```python
# let's find out the most profitable movies.

movie_and_profit = df[['original_title', 'profit']]

# Plot the data
sns.barplot(x='profit', y='original_title', data=movie_and_profit.sort_values(by='profit', ascending=False).head(10))
plt.title('Top 10 Most Profitable Movies', fontsize=16)
plt.xlabel('Profit', fontsize=14)
plt.ylabel('Movie Title', fontsize=14)
plt.show()
```

From the diagram above, we could see that the movies which generate the most amount of profit are; Avatar, Star Wars: the force awaken and Titanic.

QUESTION 2: WHAT IS THE RELATIONSHIP BETWEEN THE BUDGET AND REVENUE?

```python
#let's check the relationship between the Budget and revenue.

import matplotlib.pyplot as plt
plt.scatter(df['revenue'], df['budget'], color='blue', marker='o')
plt.title('Relationship between Revenue and Budget')
plt.xlabel('Revenue')
plt.ylabel('Budget')
plt.show()


```

From the image above, we could see that there's a positive relationship between Revenue and Budget;i.e the increase of budget amount, has a high tendency to increase the amount of revenue generated.

QUESTION 3: WHICH MOVIES ARE THE MOST POPULAR?

```python
#let's find out the movies which are most popular

Most_popular_movies = df[['original_title', 'popularity']]
sns.barplot(x = 'popularity', y= 'original_title', data=Most_popular_movies.sort_values(by = 'popularity', ascending = False).head(10))
plt.title('Most Popular Movies', fontsize = 16)
plt.xlabel('Popularity', fontsize = 14)
plt.ylabel('Movies', fontsize = 14)
plt.show()
```

Analysis shows that jurassic World, Mad Max:fury Road and intersteller are the most popular movies. 

QUESTION 4: WHAT IS THE RELARIONSHIP BETWEEN THE POPULARITY AND REVENUE OF A MOVIE?

```python
#let's find out if the popularity of movies influence its profit.
plt.scatter(df['popularity'], df['revenue'], color='blue', marker='o')
plt.title('Relationship between Popularity and Revenue')
plt.xlabel('Popularity')
plt.ylabel('Revenue')
plt.show()
```

From the image above, we could see that the relationship between populatity and profit of movies is weak.i.e the more popular the movies get, do not neccessarily increase the revenue of the movie.

QUESTION 5: WHAT IS THE PERCENTAGE OF MOVIES RELEASED PER GENRES?

```python
#let's find out the genres with the most released movies.

# Spliting the genres column, get dummy variables, and count each genre
genre = df['genres'].str.get_dummies(sep='|')

# Suming the values in each genre column to get the count of each genre
genre_counts = genre.apply(sum, axis=0)

# Sorting genres by count and plot the top 14 as a pie chart
genre_counts.sort_values(ascending=False).head(14).plot.pie(figsize=(10,10), fontsize=15, autopct='%1.1f%%')
plt.title('Movies Released Per Genre', fontsize=20)
plt.ylabel('')
plt.show()
```

from the image above, it can be seen that Drama has the highest release of movies, taking 18.7% of movies released. followed by Comedy with over 14.9% of movies released; lastly the genre with the least percentage of movies released is Documentary, with only 2.0% of movies released.
